```python
import numpy as np

def softmax(x):
    ex_fenzi= np.exp(x)
    ex_fenmu = np.sum(ex_fenzi,axis=-1,keepdims=True)

    return ex_fenzi/ex_fenmu
```

```python
x = np.array([1001,1000,999])
softmax(x)
```

两边取对数 不恰当的表示

log(ex_fenzi/ex_fenmu) = log(ex_fenzi)- log(ex_fenmu)
                       = (x-x_max) -(log(np.sum(ex_fenzi,axis=-1,keepdims=True)))

```python
import numpy as np

def log_softmax(x):
    x_max = np.max(x,axis=-1,keepdims=True)
    ex_fenzi= np.exp(x-x_max)
    ex_fenmu = np.sum(ex_fenzi,axis=-1,keepdims=True)

    return np.log(ex_fenzi)- np.log(ex_fenmu)
```

```python
x = np.array([1001,1000,999])
log_softmax(x)
```
